```python
#Changing NaN values to blank spaces for finding out the jaccard co-effient.
df = df.replace(np.nan, '', regex=True)
```

```python
def jaccard(str1, str2):
    intersection = len(list(set(str1).intersection(str2)))
    union = (len(str1) + len(str2)) - intersection
    return float(intersection) / union
```

```python
jaccard(str1, str2)
```

```python
jaccard_sim = []
for x in df.index:
    str1 = df.iloc[x,6]
    if str1 == '':
        z = 0
        jaccard_sim.append(z)
        continue
    
    str2 = df.iloc[x,14]
    if str2 == '':
        z = 0
        jaccard_sim.append(z)
        continue
    z = jaccard(str1, str2)
    jaccard_sim.append(z)
    print(z)
```

```python
len(jaccard_sim)
```

```python
#Adding Jaccard Similarity to the Data Frame
df["Jaccard_Similarity"] = jaccard_sim
```

```python
df.head()
```

```python
# Getting the index of all rows with Jaccard Similarity greater than 0.5
c = 0
for y in jaccard_sim:
    if y >= 0.5:
        print(c)
    c+=1
```
