```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('movies_metadata_update1.csv', encoding = "ISO-8859-1")
df.head()
```

```python
cols = list(df.columns.values) #Make a list of all of the columns in the df
cols.pop(cols.index('overview')) #Remove overview from list
cols.pop(cols.index('tagline')) #Remove tagline from list
df = df[cols+['overview','tagline']] #append them at the end of the dataset
```

```python
df
```

```python
#Creating train and test datasets
train = df[df['release_date'] < '20140101'] 
test = df[df['release_date'] > '20131231']
```

```python
#Replacing all non-alphabetic characters with blank spaces
data=train.iloc[:,18:20]
data.replace("[^a-zA-Z]"," ",regex=True, inplace=True)
```

```python
data
```

```python
#Converting all characters to lower case for better matching
data["overview"] = data["overview"].str.lower()
data = data.replace(np.nan, '', regex=True)
data["tagline"] = data["tagline"].str.lower()
```

```python
data
```

```python
#Joing the overview and tagline columns
' '.join(str(x) for x in data.iloc[0,0:3])
```

```python
text = []
for row in range(0,len(data.index)):
    text.append(' '.join(str(x) for x in data.iloc[row,0:3]))
```

```python
text[1]
```

```python
#The sklearn.feature_extraction module can be used to extract features in a format supported by machine learning
#algorithms from datasets consisting of formats such as text and image.

#TfidfVectorizer - Transforms text to feature vectors that can be used as input to estimator.

#CountVectorizer provides a simple way to both tokenize a collection of text documents and build a vocabulary of 
#known words, but also to encode new documents using that vocabulary
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
```

```python
#Used for Computing the distance matrix from a vector array X and optional Y.

#Cosine similarity computes similarity as the normalized dot product of X and Y
from sklearn.metrics.pairwise import linear_kernel, cosine_similarity
```

```python
#An n-gram is just a string of n words in a row.
#fit_transform is convenient and efficient for modelling and transforming the training data simultaneously.

tf = TfidfVectorizer(analyzer='word',ngram_range=(1, 2),min_df=0, stop_words='english')
tfidf_matrix = tf.fit_transform(text)
print(tfidf_matrix.shape)
```

```python
print(tfidf_matrix)
```

```python
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
```

```python
print(cosine_sim)
```
